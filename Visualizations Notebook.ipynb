```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.optimizers import RMSprop
import time as t
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('bike_rides.csv')
df = df[['power','heart-rate']]
df.dropna()
#Comment for commit

```

```python
X_train, X_test, y_train, y_test = train_test_split(df['power'], df['heart-rate'], test_size=0.2, random_state=42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
start_time = t.time()
model = Sequential([
    Dense(units=10, activation='relu', input_shape=(1,)),  # Input layer
    Dense(units=10, activation='relu'),  # Hidden layer
    Dense(units=1)  # Output layer, adjust activation
])


model.compile(optimizer=SGD(learning_rate=0.01),
              loss='mean_squared_error',
              metrics=['mean_squared_error'])  # Use 'mean_squared_error' for regression tasks


SGD_history = model.fit(X_train, y_train,
                    epochs=100,
                    batch_size=32,
                    validation_data=(X_val, y_val))


test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss}, Test MSE: {test_accuracy}")
print(f"Time taken: {t.time() - start_time}")
```
